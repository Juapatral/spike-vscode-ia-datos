```python
import pandas as pd

# 1. Definimos una función sencilla para inspeccionar los datos
# Esto ayuda a los estudiantes a ver la estructura antes de cualquier transformación
def inspeccionar_datos(df, nombre_dataset):
    print(f"=== Análisis inicial del dataset: {nombre_dataset} ===")
    print(f"1. Dimensiones (Filas, Columnas): {df.shape}")
    print(f"2. Columnas: {df.columns.tolist()}")
    print("\n3. Tipos de datos:")
    print(df.dtypes)
    print("\n4. Valores nulos por columna:")
    print(df.isnull().sum())
    print(f"\n5. Total de registros duplicados: {df.duplicated().sum()}")
    print("-" * 50, "\n")

# 2. Carga de los archivos CSV desde la carpeta 'data'
# Usamos rutas relativas tal como se solicitó
try:
    customers = pd.read_csv('../data/customers.csv')
    orders = pd.read_csv('../data/orders.csv')
    order_items = pd.read_csv('../data/order_items.csv')
    order_payments = pd.read_csv('../data/order_payments.csv')

    # 3. Ejecución de la inspección inicial solicitada
    inspeccionar_datos(customers, "customers")
    inspeccionar_datos(orders, "orders")
    inspeccionar_datos(order_items, "order_items")
    inspeccionar_datos(order_payments, "order_payments")

except FileNotFoundError as e:
    print(f"Error: Asegúrate de que los archivos CSV estén en la carpeta 'data'. Detalle: {e}")
```

```python
tables = {
    "customers": customers,
    "orders": orders,
    "items": order_items,
    "payments": order_payments,
}

expected_columns = {
    "customers": {"customer_id", "customer_unique_id", "customer_city", "customer_department"},
    "orders": {"order_id", "customer_id", "order_status", "order_purchase_timestamp"},
    "items": {"order_id", "order_item_id", "product_category", "price", "freight_value"},
    "payments": {"order_id", "payment_sequential", "payment_type", "payment_value"},
}

for name, expected in expected_columns.items():
    missing = expected - set(tables[name].columns)
    assert not missing, f"{name} no tiene columnas esperadas: {missing}"
```

```python
import pandas as pd

def add_date_columns(df, date_column, prefix=''):
    """
    Genera las columnas day, month, year a partir de una columna de fecha.
    Se puede especificar un prefijo para los nombres de las nuevas columnas.
    """
    df[date_column] = pd.to_datetime(df[date_column])
    
    col_prefix = f"{prefix}_" if prefix else ""
    
    df[f'{col_prefix}day'] = df[date_column].dt.day
    df[f'{col_prefix}month'] = df[date_column].dt.month
    df[f'{col_prefix}year'] = df[date_column].dt.year
    
    return df
```

```python
add_date_columns(orders, 'order_purchase_timestamp', prefix='purchase_date')
```
